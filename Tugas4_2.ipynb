**Cleaning**

```
import pandas as pd
import numpy as np
```

```
path = "/content/drive/My Drive/Kuliah/Data Mining/Tugas 4/titanic3.xls"
df = pd.read_excel(path)
```

```
df.head()
```

```
df.info()
```

```
replace_missing_values = {"age" : 0.0, "fare": 0.0, "embarked":"unknown"}
```

```
df = df.fillna(replace_missing_values)
```

```
df
```

```
df.info()
```

```
df['cabin']=df.cabin.apply(lambda x:0 if type(x)==float else 1)
```

```
df['boat']=df.boat.apply(lambda x:0 if type(x)==float else 1)
```

```
df['sex'] = df['sex'].map({'female':0, 'male':1})
```

```
df['age']= round(df['age'], 0)
```

```
df['age'] = df['age'].astype(int)
```

```
df['fare'] = round(df['fare'], 0)
```

```
df['fare'] = df['fare'].astype(int)
```

```
df['embarked'] = df['embarked'].map({'S':1,'C':2,'unknown':0,'Q':3})
```

```
df = df.drop(['name', 'ticket', 'home.dest' ], axis=1)
```

```
df
```

```
df_corr = df.corr()["survived"]
df_corr.abs().sort_values(ascending=False)[1:]
```

**KNN**

```
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```
X = df.filter(['boat','sex','pclass','cabin','fare'], axis=1)
Y = df['survived']
```

```
X_train, X_test,Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

```
classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
classifier.fit(X_train, Y_train)
```

```
Y_pred= classifier.predict(X_test)
```

```
print(metrics.accuracy_score(Y_test, Y_pred))
```

```
print(confusion_matrix(Y_test, Y_pred))
```

```
print(classification_report(Y_test, Y_pred))
```
